```python
# Import necessary libraries
import numpy as np

# pyplot is for general plotting
import matplotlib.pyplot as plt

# wavfile is for reading flute.wav
from scipy.io import wavfile
```

Let's create some data. Here we're making a signal consisting of 2 sine waves (1250 $Hz$ and 625 $Hz$) sampled at a 10 $kHz$ rate.

```python
x = np.arange(256.0)
sin1 = np.sin(2*np.pi*(1250.0/10000.0)*x)
sin2 = np.sin(2*np.pi*(625.0/10000.0)*x)
sig = sin1 + sin2
```

What does this look like? Let's plot it and find out.

```python
plt.plot(sig)
```

Notice with IPython Notebook we don't have to clear our plots with `clf()`! Also we don't need to define a plotting/graphing function since plotting in line is super easy!

Our first plot showed the signal in the time domain. We want to see it in the frequency domain. A numpy function that implements an algorithm called the Fast Fourier Transform (FTT) can take us there.

```python
data = np.fft.rfft(sig)
plt.plot(data)
```

That's a start. We had 2 frequencies in the signal, and we're seeing 2 spikes here, so that seems reasonable.

We got a warning because the fourier transform gave us a complex output - so instead we need to take the magnitude of the complex output. [Why is this no longer generating a warning? We *should* get a warning.]

```python
data = abs(data)
plt.plot(data)
```

But this is displaying raw power output, and we usually think of audio volume in terms of decibles. Wikipedia tells us decibels (dB) are the original signal plotted on a 10*log10 y-axis, so:

```python
data = 10*np.log10(data)
plt.plot(data)
```

We see our 2 pure tones showing up as 2 peaks. This is great! The jaggedness of the rest of the signal is quantization noise, i.e. numerical error, because we're doing this with approximations.

What's the relationship of the x-axis of the graph and the frequency of the signal?

Well, the numpy fft function goes from 0-5000 $Hz$ by default. This means the x-axis markers correspond to values of 0-5000 $Hz$ divided into 128 slices. 

$5000/128\ =\ 39.0625\ Hz$ per marker

The two peaks are at 16 and 32.

$$(5000/128)*16\ {\approx}\ 625\ Hz$$  
$$(5000/128)*32\ {\approx}\ 1250\ Hz$$

which are our 2 original tones!

Let's generate and plot a spectrogram.

```python
sgram = plt.specgram(sig)
```

The spectrogram is sort of like our last plot, extruded forward out of the screen, and looked upon from above. Time is on the x-axis, frequency on the y-axis, and amlitude is marked by color.

Now let's do this with a more complex sound. There should be a file called 'flute.wav' in the same directory as this notebook to execute this code.

```python
# wavfile.read returns a tuple of length two where the first
# entry is the sample rate and the second is the audio data
sampling_rate, audio = wavfile.read('flute.wav')
```

Let's make sure we have the correct data by writing back to a .wav file and making sure they sound the same.

```python
wavfile.write('reflute.wav', sampling_rate, audio)
```

Now let's see what this looks like in the time domain.

```python
plt.plot(audio)
```

We've got a lot of data points, so we'll only plot the beginning of the signal here.

```python
plt.plot(audio[:1024])
```

What does this look like in the frequency domain?

```python
audio_fft = np.fft.rfft(audio)
audio_fft = abs(audio_fft)
audio_fft = 10*np.log10(audio_fft)
plt.plot(audio_fft)
```

This is much more complex than our last example. We can see harmonics on the left side. Perhaps this will be clearer if we plot it as a spectrogram.

```python
audio_sgram = plt.specgram(audio)
```

You can see the base note of the flute (a 494 $Hz$ B) in dark red at the bottom, and lighter red harmonics above it.
